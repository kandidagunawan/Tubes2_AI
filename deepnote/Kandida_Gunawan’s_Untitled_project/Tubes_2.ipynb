## KNN AND NAIVE BAYES USING SCIKIT-LEARN

### KNN USING SCIKIT-LEARN

```
y = data_train['price_range']
X = data_train.drop('price_range', axis = 1)
knnModel = KNeighborsClassifier(n_neighbors = 1)
knnModel.fit(X, y)

y_true_test = data_validation['price_range']
X_test = data_validation.drop('price_range', axis = 1)
y_pred_test = knnModel.predict(X_test)
accuracy_score1 = accuracy_score(y_true_test, y_pred_test)
print(f'Accuracy Score of KNN Algorithm : {accuracy_score1}')
```

### NAIVE BAYES USING SCIKIT-LEARN

```
y = data_train['price_range']
X = data_train.drop('price_range', axis = 1)
naiveBayesModel = GaussianNB()
naiveBayesModel.fit(X, y)

y_true_test = data_validation['price_range']
X_test = data_validation.drop('price_range', axis = 1)
y_pred_test = naiveBayesModel.predict(X_test)
accuracy_score2 = accuracy_score(y_true_test, y_pred_test)
print(f'Accuracy Score of Naive Bayes Algorithm : {accuracy_score2}')
```
