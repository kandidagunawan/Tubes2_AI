```
# Box plot of every column
numerical_columns = ['battery_power', 'clock_speed', 'fc', 'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time']
for column in numerical_columns:
  print(column)
  q1, q3 = np.percentile(train_set[column], [25, 75])
  iqr = q3-q1
  print(f'q1: {q1}')
  print(f'q3: {q3}')
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  print(f'lower bound: {lower_bound}')
  print(f'upper bound: {upper_bound}')
  plt.title(column)
  plt.boxplot(train_set[column])
  fig = plt.figure(figsize =(10, 7))

  plt.show()

```

## DISTRIBUSI DATA KOLOM NUMERIK

```
# HISTOGRAM
numerical_columns = ['battery_power', 'clock_speed', 'fc', 'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time']
for col in numerical_columns:
  plt.figure(figsize = (8, 4))
  plt.hist(train_set[col], color='skyblue', edgecolor='black')
  plt.title(f'{col} \'s Histogram')
  plt.xlabel(col)
  plt.ylabel("Frequency")
  plt.show()
```

```
def normalityTest(data, column_name, row, col, index):
    k, p = stats.normaltest(data)
    alpha = 0.05
    mean, std = stats.norm.fit(data)
    plt.subplot(row, col, index)
    plt.hist(data, bins='auto', density=True)
    xmin, xmax = plt.xlim()
    x = np.linspace(xmin, xmax, 100)
    y = stats.norm.pdf(x, mean, std)
    plt.plot(x, y, 'k', lw=1)
    plt.title(f'{column_name} Distribution')
    plt.xlabel(column_name)
    plt.ylabel('Density')
    if p < alpha:
        plt.text(0.1, 0.9, f"p = {p:.4f}\nNot normally distributed", transform=plt.gca().transAxes, fontsize=10, color='red')
    else:
        plt.text(0.1, 0.9, f"p = {p:.4f}\nNormally distributed", transform=plt.gca().transAxes, fontsize=10, color='green')

numerical_columns = ['battery_power', 'clock_speed', 'fc', 'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time']
num_rows = 4
num_cols = 4
fig, axs = plt.subplots(num_rows, num_cols, figsize=(15, 15))
axs = axs.ravel()

for i, col in enumerate(numerical_columns):
    normalityTest(train_set[col], col, num_rows, num_cols, i + 1)
plt.tight_layout()
plt.show()
```

```
# Kurtosis
numerical_columns = ['battery_power', 'clock_speed', 'fc', 'int_memory', 'm_dep', 'mobile_wt', 'n_cores', 'pc', 'px_height', 'px_width', 'ram', 'sc_h', 'sc_w', 'talk_time']

kurtosis_values = [stats.kurtosis(train_set[col]) for col in numerical_columns]
plt.figure(figsize=(10, 6))
plt.plot(numerical_columns, kurtosis_values, marker='o', color='skyblue', linestyle='-', markersize=8, markeredgecolor='black')
plt.xlabel('Numerical Columns')
plt.ylabel('Kurtosis')
plt.title('Kurtosis for Numerical Columns')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

data =[]

for col in numerical_columns:
    kurtosis_value = stats.kurtosis(train_set[col])
    if kurtosis_value > 0:
        kurtosis_desc = "Leptokurtic"
    elif kurtosis_value < 0:
        kurtosis_desc = "Platykurtic"
    else:
        kurtosis_desc = "Mesokurtic"
    data.append({"Column Name": col, "Kurtosis Value": kurtosis_value, "Kurtosis Desc": kurtosis_desc})
kurtosis = pd.DataFrame(data)
kurtosis
```

## BAR PLOT UNTUK KOLOM NON NUMERIK

```
# Bar Plot
non_numerical_columns = ['blue', 'dual_sim', 'four_g', 'three_g', 'touch_screen', 'wifi', 'price_range']



num_columns = len(non_numerical_columns)

# Create a single figure with subplots
fig, axes = plt.subplots(1, num_columns, figsize=(16, 4))

for i, col in enumerate(non_numerical_columns):
    col_counts = train_set[col].value_counts()

    # Create a bar plot in the i-th subplot
    col_counts.plot(kind='bar', color='skyblue', ax=axes[i])

    axes[i].set_title(f'Bar Plot of {col}')
    axes[i].set_xlabel('Categories')
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=0)  # Rotate x-axis labels if needed

plt.tight_layout()  # Adjust subplot layout to prevent overlap
plt.show()

```
